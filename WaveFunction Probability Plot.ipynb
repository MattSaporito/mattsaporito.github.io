```python
import numpy as np
import matplotlib.pyplot as plt

# Given constants
gamma = 2  # per nm
a = 0.3  # nm

# Define the wavefunction
def psi(x, A):
    return A * (np.exp(-gamma * np.abs(x + a)) - np.exp(-gamma * np.abs(x - a)))

# Define the probability density function
def psi_squared(x, A):
    return np.abs(psi(x, A))**2

# Generate x values
x_values = np.linspace(-2, 2, 1000)  # from -2 nm to 2 nm

# Since the wavefunction is normalized, A = sqrt(2*gamma)
A_value = np.sqrt(2 * gamma)

# Calculate wavefunction and probability values
psi_values = psi(x_values, A_value)
probability_values = psi_squared(x_values, A_value)

# Plotting
plt.figure(figsize=(12, 6))

# Plot wavefunction
plt.subplot(1, 2, 1)
plt.plot(x_values, psi_values, label='$\psi(x)$', color='blue')
plt.title('Wavefunction $\psi(x)$')
plt.xlabel('x (nm)')
plt.ylabel('$\psi(x)$')
plt.grid(True)
plt.legend()

# Plot probability density
plt.subplot(1, 2, 2)
plt.plot(x_values, probability_values, label='$|\psi(x)|^2$', color='red')
plt.title('Probability Density $|\psi(x)|^2$')
plt.xlabel('x (nm)')
plt.ylabel('$|\psi(x)|^2$')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
