```python
import numpy as np
import matplotlib.pyplot as plt

π = np.pi
```

```python
Δt = 0.1 #s
t0 = 0 #s
tf = 10.0 #s

time = np.arange(t0, tf + Δt, Δt)

print(time)
```

#   <font color='Red'> **Experiment 1** </font>

### <font color='Red'> To begin lets start with by defining an initial position. </font>

```python
x_0=np.array([1., 0., 0.,]) #m
print (x_0)
```

### <font color='Red'> PRINT the distance of the previously defined position vector (give units!): </font>

```python
distance = np.linalg.norm(x_0)

print("The initial distance is %4.3f meters "  % distance)
```

### <font color='Red'> Define a for-loop to Velocity-Verlet integrate the motion to obtain a x_t. </font>

```python
x_t = np.zeros((len(time), 3))
v_t = np.zeros((len(time), 3))

x_t[0] = x_0
for t in range(1,len(time)):
  x_t[t] = x_t[t-1] + v_t[t-1] * Δt
  v_t[t] = v_t[t-1]
```

### <font color='Red'> Take a time-derivative of x_t to obtain dxdt_t. </font>

```python
dxdt_t = np.gradient(x_t, axis=0)/Δt
```

#   <font color='Red'> **Experiment 3** </font>

###  <font color='Red'> To begin lets start with by defining an initial position, velocity, and acceleration </font>

```python
x_0 = np.array([1., 0., 0.]) # m
v_0 = np.array([1., 0., 0.]) # m/s
a_0 = np.array([2., 0., 0.]) # m/s/s
```

###  <font color='Red'> PRINT out the distance, speed, and magnitude of the acceleration of the previously defined vectors: </font>

```python
distance = np.linalg.norm(x_0)
speed = np.linalg.norm(v_0)
acceleration = np.linalg.norm(a_0)

print("the initial distance is %4.3f meters "  % distance)
print("the initial speed is %4.3f meters/second "  % speed)
print("the initial acceleration is %4.3f meters/second/second "  % acceleration)
```

###  <font color='Red'> Define a for-loop to Velocity-Verlet integrate the motion to obtain a x_t, v_t, and a_t. </font>

```python
x_t = np.zeros((len(time), 3))
v_t = np.zeros((len(time), 3))
a_t = np.zeros((len(time), 3))

x_t[0] = x_0
v_t[0] = v_0
a_t[0] = a_0
for t in range(1,len(x_t)):
  x_t[t] = v_t[t-1]  + 0.5 * a_t[t-1]^2 + x_t[t]
  v_t[t] = v_t[t-1] + 0.5 * ( a_t[t-1] + a_t[t] ) * Δt
  a_t[t] = a_t[t-1]
```

###  <font color='Red'> Take a time-derivative of x_t to obtain dxdt_t. </font>

###  <font color='Red'> Take a time-derivative of v_t to obtain dvdt_t. </font>

###  <font color='Red'> Take a time-derivative of a_t to obtain dadt_t. </font>

```python
dxdt_t = np.gradient(x_t, axis=0)/Δt
dvdt_t = np.gradient(v_t, axis=0)/Δt
dadt_t = np.gradient(a_t, axis=0)/Δt
```

###  <font color='Red'> Make 4 plots of with: x_t, \{v_t, dxdt_t\}, \{ a_t, dvdt_t\}, and dadt_t in matplotlib, all as a function of time. Make sure to include all LABELS!. </font>

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
fig.suptitle("Experiment 3")
fig.subplots_adjust(hspace=0.01, wspace=0.25, right=2)

ax1.scatter(time, x_t[:,0])
ax1.set_title("Distance vs Time")
ax1.legend()
ax1.set(xlabel="Time (s)", ylabel="Distance from Origin (m)")

ax2.scatter(time, np.linalg.norm(v_t   , axis=1), label="|v|")
ax2.scatter(time, np.linalg.norm(dxdt_t, axis=1), s=15, label="|dx/dt|")
ax2.set_title("Speed vs Time")
ax2.legend()
ax2.set(xlabel="Time (s)", ylabel="Speed (m/s)")

ax3.scatter(time, a_t[:,0], label="Acceleration")
ax3.scatter(time, dvdt_t[:,0], s=0.75, label="dv/dt")
ax3.legend(loc="upper right")
ax3.set_title("Acceleration vs Time")
ax3.set(xlabel="Time (s)", ylabel="Acceleration (m/s/s)")

ax4.scatter(time, np.linalg.norm(dadt_t, axis=1), s=15, label="|da/dt|")
ax4.set_ylim([0, 4])
ax4.set_title("Jerk vs Time")
ax4.set(xlabel="Time (s)", ylabel="Jerk (m/s$^3$)")
ax4.legend()
fig.show()
```

#   <font color='Red'> **Experiment 4** </font>

###   <font color='Red'> To begin lets start given the following initial position, velocity, and acceleration, WITH velocity and acceleration in different directions! </font>

```python
import numpy as np

Δt = 0.15
time = np.arange (0, 20, Δt)

x_0 = np.array([3., 0., 0.]) # m
v_0 = np.array([0., 100., 0.]) # m/s
a_0 = np.array([0., -9.81, 0.]) # m/s/s
```

###  <font color='Red'> Define a for-loop to Velocity-Verlet integrate the motion to obtain a x_t, v_t, and a_t. </font>

```python
x_t = np.zeros((len(time), 3))
v_t = np.zeros((len(time), 3))
a_t = np.zeros((len(time), 3))

x_t[0] = x_0
v_t[0] = v_0
a_t[0] = a_0
for t in range(1, len(time)):
  x_t[t] = x_t[t-1] + v_t[t-1] * Δt + a_t[t-1] * Δt*Δt/2
  a_t[t] = a_t[t-1]
  v_t[t] = v_t[t-1] + (a_t[t] + a_t[t-1])/2  * Δt
```

###  <font color='Red'> Take a time-derivative of x_t to obtain dxdt_t. </font>

###  <font color='Red'> Take a time-derivative of v_t to obtain dvdt_t. </font>

###  <font color='Red'> Take a time-derivative of a_t to obtain dadt_t. </font>

```python
dxdt_t = np.gradient(x_t, axis=0)/Δt
dvdt_t = np.gradient(v_t, axis=0)/Δt
dadt_t = np.gradient(a_t, axis=0)/Δt
```

###  <font color='Red'> Make 4 plots of with: x_t, \{v_t, dxdt_t\}, \{ a_t, dvdt_t\}, and dadt_t in matplotlib, all as a function of time. Make sure to include all LABELS!. </font>

```python
plt.plot(time, x_t)
plt.set_title("Distance")
plt.legend()
plt.set(xlabel="time (s)", ylabel="Distance (m)")
plt.show()

plt.plot(time, v_t)
plt.show()

plt.plot(time, a_t)
plt.show()
```

1.5.1 Concept of Inertia

Inertia is the tendency of an object to resist change in its current state of motion. In other words, objects want to keep doing what they’re doing. An object will continue to stay at rest or keep moving until an outside force acts upon it. In this lab, we simulated this scenario in Experiment 1. The object stayed at rest with no change in position and hence no force acted upon it. In Experiment 2, the object continued to move with a constant velocity since there was no outside force acting upon it. In Experiments 3 and 4, an external force was present and thus a constant acceleration. In these cases, the object’s inertia was overcome and both position and speed changed with time. 

1.5.2 The Velocity-Verlet Integrator

The Velocity-Verlet Integrator is a numerical method and programming algorithm that is used to determine the solution of the kinematic equation for an object’s motion. As it applies to this simulated experiment, it integrates the object’s motion to obtain position, velocity, and acceleration with respect to time. Therefore, starting with an equation of an object’s motion, it is possible to predict the future states of the object as it moves using the Velocity-Verlet Integrator.
